```python
#basic packages
import numpy as np
import pandas as pd
```

```python
dfCombined = pd.read_csv('data/Combined_News_DJIA_Prices.csv')
```

```python
def create_TPlusN_Label(df,window):    
    length=df['Label'].shape[0]
    result=[]
    for index in range(0,length):
        if index >= length - window : 
            result.append(None)
        else:
            result.append(df['Label'][index+window])
    return np.array(result)    
```

```python
def create_TPlusN_Close(df,window):    
    length=df['Close'].shape[0]
    result=[]
    for index in range(0,length):
        if index >= length - window : 
            result.append(None)
        else:
            result.append(df['Close'][index+window])
    return np.array(result)    
```

```python
dfCombined['LabelTPlus5']=create_TPlusN_Label(dfCombined,5)
dfCombined['CloseTPlus5']=create_TPlusN_Close(dfCombined,5)
dfCombined['LabelTPlus20']=create_TPlusN_Label(dfCombined,20)
dfCombined['CloseTPlus20']=create_TPlusN_Close(dfCombined,20)
dfCombined['LabelTPlus60']=create_TPlusN_Label(dfCombined,60)
dfCombined['CloseTPlus60']=create_TPlusN_Close(dfCombined,60)
```
