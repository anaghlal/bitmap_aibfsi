```python
#basic packages
import numpy as np
import pandas as pd
```

```python
def direction(price_tplusone, price_t):
    if price_tplusone > price_t :
        return 1
    return 0
```

```python
df = pd.read_csv('data/DJIA_table.csv')
df["date_obj"] = pd.to_datetime(df["Date"])
df = df.sort_values(by="date_obj",ignore_index="true")
df['Label'] = df['Close']
for index in df.index:
    if index == 0:
            df.at[index,'Label']=None
    else :
        df.at[index,'Label']= direction(df['Close'][index],df['Close'][index-1])
        
        
```

```python
def create_TPlusN(df,N,srcCol='Label',colName='Label5'):    
    length = df.shape[0]
    df[colName]=df[srcCol]
    for index in df.index:
        if index >= length - N : 
            df.at[index,colName]=None
        else:
            df.at[index,colName] = df[srcCol][index+N]
    return df    
```

```python
df = create_TPlusN(df,5,'Label','Label5')
df = create_TPlusN(df,5,'Close','Close5')
df = create_TPlusN(df,20,'Label','Label20')
df = create_TPlusN(df,20,'Close','Close20')
df = create_TPlusN(df,60,'Label','Label60')
df = create_TPlusN(df,60,'Close','Close60')
```
